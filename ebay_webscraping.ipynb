```python
# full list Sponsor
full_sponsored = []
for pgn in range(1,11):
    filename = 'Ipad' + str(pgn).zfill(2) + '.htm'
    with open(filename, 'r') as file:
        text = file.read()
        soup = BeautifulSoup(text, 'html.parser') 
        items = soup.select("#srp-river-results > ul > li.s-item")
            
        #list
        sponsored = []
        for item in items:
            a = item.find("a")
            if not a:
                continue
                
            text = item.find("span").get_text()
            if re.search('.*S.*P.*O.*N.*S.*O.*R.*E.*D.*',text):  
                    #print(a['href'])
                url_simplify = re.sub("(.*)\\?.*", r"\1", a['href'])           
                #print(url_simplify)
                    # apeend list
                    
                sponsored.append(url_simplify)
    # extend full list
    full_sponsored.extend(sponsored)
```

```python
for item in full_sponsored:
    with open("sponsored.txt", 'a', encoding = 'utf-8') as file:
        file.write(item + "\n")
```

```python
####NON SPONSORED

# full list
full_nonsponsored = []
for pgn in range(1,11):
    filename = 'Ipad' + str(pgn).zfill(2) + '.htm'
    with open(filename, 'r') as file:
        text = file.read()
        soup = BeautifulSoup(text, 'html.parser') 
        items = soup.select("#srp-river-results > ul > li.s-item")
            
        #list
        nonsponsored = []
        for item in items:
            a = item.find("a")
            if not a:
                continue
                
            text = item.find("span").get_text()
            if not re.search('.*S.*P.*O.*N.*S.*O.*R.*E.*D.*',text):  
                    #print(a['href'])
                url_simplify = re.sub("(.*)\\?.*", r"\1", a['href'])           
                #print(url_simplify)
                    # apeend list
                    
                nonsponsored.append(url_simplify)
    # extend full list
    full_nonsponsored.extend(nonsponsored)
```

```python
for item in full_nonsponsored:
    with open("non-sponsored.txt", 'a', encoding = 'utf-8') as file:
        file.write(item + "\n")
```

```python
###D 
item = []
items = []

folders = ['./sponsored/', './non-sponsored/']
for folder in folders:
    sponsored_dictionary = folder
    sponsored_list = os.listdir(sponsored_dictionary)
    for file in sponsored_list:
        if not file.endswith('htm'):
            continue
        fname = os.path.join(sponsored_dictionary,file)
        html_file_content = open(fname, 'r', encoding = 'utf-8').read()
        soup_sponsored = BeautifulSoup(html_file_content, 'html.parser')
        try:
            seller = soup_sponsored.find('div',{'class':'mbg vi-VR-margBtm3'}).find('span',{'class':'mbg-nw'}).text
            item.append(seller)
        except:
            seller = None
            item.append(seller)
        
        #Seller Score
        score = re.sub('\D', '', soup_sponsored.find("span","mbg-l").text)
        item.append(score)
        
        ##PRICE
        if not soup_sponsored.find('span', {'id': 'prcIsum'}):
            item_price = soup_sponsored.find('span', {'id': 'mm-saleDscPrc'}).attrs['content']
        else: 
            item_price = soup_sponsored.find('span', {'id': 'prcIsum'}).attrs['content']
        item.append(item_price)
        
        ##Best Offer
        if not soup_sponsored.find('div', {'class': 'vi-bbox-dspn u-flL lable boLable'}):
            best_offer = 'NO'
        else: best_offer = 'YES'
        item.append(best_offer)

        ##Title
        title = soup_sponsored.find('h1', {'class': 'it-ttl'}).text
        title_split = re.split(r'\s{2,}', title)[1]
        item.append(title_split)

        ##Return allowed
        returns_txt = soup_sponsored.find('span', {'id': 'vi-ret-accrd-txt'}).text
        if returns_txt == 'Seller does not accept returns':
            returns = 'NO'
        else:
            returns = 'YES'
        item.append(returns)

        ##Shipping List
        if not soup_sponsored.find('input', {'id': 'shGetRates'}):
            shipping_computed = 'NO'
        else:
            shipping_computed = 'YES'
        item.append(shipping_computed)

        ##Condition
        condition = soup_sponsored.find('div', {'id': 'vi-itm-cond'}).text
        item.append(condition)

        ##Sponsor yes vs no
        if folder == './sponsored/':   
            sponsor = 'YES'
        else:
            sponsor = 'NO' 
        item.append(sponsor)
        items.append(item)
        item = []
```
